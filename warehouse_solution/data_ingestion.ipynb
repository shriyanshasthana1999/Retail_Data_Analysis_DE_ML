```python
import pandas as pd
from pathlib import Path
import datetime
import shutil
```

Create connection to SSMS

```python
from sqlalchemy import create_engine
import urllib

params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-D3LC2OB\SQLSERVER2019;"
    "DATABASE=retail_db;"
    "Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
print("engine created successfully")
```

Alternate way to create connection to SSMS

```python
#### Alternate way to create connection to SSMS ####
##This method will create the database if it does not exist, otherwise it will continue##
import pyodbc

conn = pyodbc.connect('Driver={SQL Server};' 'Server=DESKTOP-D3LC2OB\SQLSERVER2019;' 'Database=test_db;' 'Trusted_Connection=yes;')     
print("Connected Successfully") 

cursor = conn.cursor()
print("cursor created")

cursor.execute("""
IF NOT EXISTS (
    SELECT name 
    FROM sys.databases 
    WHERE name = 'retail_db'
)
CREATE DATABASE retail_db
""")

```

Get The List of Files from the present Directory

```python
import os
files_list = os.listdir("..") #.. stands for present working directory, use the files path here
print(files_list)
```

Create Tables for Files and Archive

```python
processed_folder = '..\processed'
os.makedirs(processed_folder,exist_ok=True) #create the archived folder

currentdatetime = datetime.datetime.now().strftime("%Y%m%d%H%M%S") #get the current process timestamp

for file in files_list:
    if file.endswith(".csv"):  #check for csv files, cases like excels can also be handled
        file_df = pd.read_csv(f"..\{file}") #read the csv files as df
        table_name = "raw_" + Path(file).stem #adding raw prefix for source data 
        print(table_name)
        file_df.to_sql(  #create table and insert data into sql server, append mode is used to avoid overwriting the existing data
            name = table_name,
            con=engine,
            if_exists='append',
            index=False
        )
        shutil.move(f"..\{file}",f"{processed_folder}\{file}_{currentdatetime}") 

        
```
